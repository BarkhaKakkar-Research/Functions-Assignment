```python
#Theory Questions
'''
1. What is the difference between a function and a method in Python?
Ans. 1. Definition

A function in Python is a block of reusable code that performs a specific task and is defined independently
using the def keyword. It does not belong to any class and can be called directly by its name.
A method, on the other hand, is a function that is defined inside a class and is associated with an object. Methods are
used to define the behavior of objects created from a class.

Example:
# Function
def greet():
    print("Hello")

greet()

# Method
class Person:
    def greet(self):
        print("Hello")

p = Person()
p.greet()

2. Association with Class or Object

A function exists independently and can be used anywhere in the program without being tied to a class.
A method is always associated with a class and operates on the data (attributes) of the object created from that class.

Key Points:

Function - Independent

Method → Belongs to a class

Example:

# Function
def square(n):
    return n * n

# Method
class Math:
    def square(self, n):
        return n * n

3. Calling Style

Functions are called directly using their function name followed by parentheses.
Methods are called using an object name followed by a dot (.) and the method name.

Example:

# Calling a function
print(square(5))

# Calling a method
m = Math()
print(m.square(5))

4. Use of self Parameter

Functions do not use the self parameter because they are not associated with any object.
Methods use self as the first parameter, which refers to the current object and allows access to the object’s data and other methods.

Example:

class Student:
    def show(self):
        print("This is a method using self")

5. Purpose and Usage

Functions are mainly used for general-purpose tasks, such as calculations, printing values, or processing data.
Methods are used to define the behavior of objects in object-oriented programming.

Bullet Comparison:

Functions → General tasks

Methods → Object behavior

Example:

# Function
def add(a, b):
    return a + b

# Method
class Calculator:
    def add(self, a, b):
        return a + b

6. Real-Life Analogy

A function is like a standalone machine that can be used anywhere.
A method is like a feature of a device, which can only be used through that device.

Example:

# Function
len("Python")

# Method
"Python".upper()

Conclusion:
A function is independent and can exist on its own.
A method is a function that belongs to a class and works on object data.
Every method is a function, but not every function is a method.


2. Explain the concept of function arguments and parameters in Python.
Ans. 1. Meaning of Parameters
Parameters are the variables listed inside the parentheses in a function definition. They act as
placeholders that receive values when the function is called. Parameters allow a function to work with
different values without changing the function code.

Key Points:
Defined while creating a function
Act as local variables inside the function

Example:
def add(a, b):   # a and b are parameters
    return a + b

2. Meaning of Arguments
Arguments are the actual values passed to the function when it is called. These values are
assigned to the corresponding parameters during function execution.

Key Points:
Provided when calling a function
Supply real data to the function

Example:
result = add(5, 3)   # 5 and 3 are arguments

3. Difference Between Parameters and Arguments
Parameters are part of the function definition, while arguments are part of the function call.
Parameters receive the data, whereas arguments send the data to the function.

Example:
def multiply(x, y):   # x, y → parameters
    return x * y
multiply(4, 6)        # 4, 6 → arguments

4. Types of Function Arguments in Python
Python supports different types of arguments to provide flexibility while calling functions.

a) Positional Arguments
In positional arguments, the values are passed in the same order as the parameters are defined. The position of arguments matters.
Example:
def subtract(a, b):
    return a - b
print(subtract(10, 4))

b) Keyword Arguments
In keyword arguments, values are passed using the parameter names. Order does not matter.
Example:
def student(name, age):
    print(name, age)
student(age=20, name="Riya")

c) Default Arguments
Default arguments have a predefined value. If no argument is provided during the function call, the default value is used.
Example:
def greet(name="Student"):
    print("Hello", name)
greet()
greet("Amit")

d) Variable-Length Arguments
Sometimes, the number of arguments is unknown. Python allows this using special symbols.
*• args (Non-keyword variable-length arguments):
def total(*numbers):
    return sum(numbers)
print(total(1, 2, 3, 4))
**• kwargs (Keyword variable-length arguments):
def details(**info):
    for key, value in info.items():
        print(key, value)
details(name="Anita", age=22)

5. Importance of Arguments and Parameters
Arguments and parameters make functions reusable, flexible, and modular. They allow the
same function to work with different data, reducing code repetition.

Example:
def square(n):
    return n * n
print(square(5))
print(square(8))


3.What are the different ways to define and call a function in Python?
Ans. In Python, a function can be defined and called in multiple ways depending on the type of arguments
used and the programming requirement. These methods make Python functions flexible and reusable.

1. Defining and Calling a Simple Function (No Arguments)
A function can be defined without any parameters. Such functions perform a fixed task and are
called using only the function name followed by parentheses.
Key Points:
No parameters
Same output every time

Example:
def greet():
    print("Hello, Welcome to Python")
greet()

2. Function with Parameters and Arguments
A function can accept parameters to receive values during execution. The function is called by passing arguments.
Key Points:
Parameters are defined in function definition
Arguments are passed during function call

Example:
def add(a, b):
    return a + b
result = add(4, 6)
print(result)

3. Function with Positional Arguments
In positional arguments, values are passed in the same order as the parameters are defined.
Key Points:
Order matters
Most commonly used

Example:
def divide(a, b):
    print(a / b)
divide(10, 2)

4. Function with Keyword Arguments
Keyword arguments allow passing values using parameter names. The order of arguments does not matter.
Key Points:
More readable
Order independent

Example:
def student(name, age):
    print(name, age)
student(age=18, name="Ravi")

5. Function with Default Arguments
Default arguments have predefined values. If no value is passed, the default value is used.
Key Points:
Avoids errors when arguments are missing
Increases flexibility

Example:
def greet(name="Student"):
    print("Hello", name)
greet()
greet("Neha")

6. Function with Variable-Length Arguments
When the number of arguments is unknown, Python provides variable-length arguments.
a) *args (Non-keyword arguments)
Allows passing multiple values as a tuple.

Example:
def total(*numbers):
    print(sum(numbers))
total(2, 4, 6, 8)

b) **kwargs (Keyword arguments)
Allows passing multiple key–value pairs as a dictionary.

Example:
def details(**info):
    for key, value in info.items():
        print(key, value)
details(name="Asha", age=20)

7. Function with Return Statement
Functions can return values using the return keyword. The returned value can be stored or used in expressions.
Key Points:
Ends function execution
Sends result back to caller

Example:
def square(n):
    return n * n
print(square(5))

8. Lambda (Anonymous) Function
A lambda function is a small, one-line function without a name. It is defined using the lambda keyword.
Key Points:
Short and concise
Used for simple operations

Example:
square = lambda x: x * x
print(square(6))


4. What is the purpose of the `return` statement in a Python function?
Ans. 1. Returning a Value from a Function
The primary purpose of the return statement is to send a value back to the function caller.
When a function performs a computation, the result can be returned and used elsewhere in the program.

Key Points:
Sends output to the calling code
Makes functions useful and reusable

Example:
def add(a, b):
    return a + b
result = add(3, 5)
print(result)

2. Ending the Execution of a Function
The return statement immediately terminates the execution of a function. Any code written
after the return statement inside the function will not be executed.

Key Points:
Stops function execution
Control goes back to the caller

Example:
def check_number(n):
    if n > 0:
        return "Positive"
    return "Non-positive"

3. Returning Multiple Values
In Python, a function can return multiple values using a single return statement. These values are returned as a tuple.

Key Points:
Multiple outputs possible
Returned as a tuple

Example:
def calculate(a, b):
    return a + b, a - b
sum_val, diff_val = calculate(10, 5)
print(sum_val, diff_val)

4. Returning No Value (None)
If a function does not include a return statement, or if return is used without an expression, the function returns None by default.

Key Points:
None indicates no value
Useful for action-only functions

Example:
def display():
    print("Hello")
result = display()
print(result)   # Output: None

5. Using return in Conditional Statements
The return statement is often used inside conditional blocks to return different results based on conditions.

Key Points:
Improves decision-making
Avoids unnecessary code execution

Example:
def even_or_odd(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

6. Difference Between return and print
The return statement sends a value back to the caller, while print only displays the value on the screen.

Example:
def square(n):
    return n * n
print(square(4))   # Value can be reused


5.What are iterators in Python and how do they differ from iterables?
Ans. 1. What is an Iterable?
An iterable is any Python object that can be looped over (iterated) one element at a time. Iterables store multiple
values and can return an iterator using the iter() function. Common examples of iterables include lists, tuples, strings, sets,
and dictionaries.

Key Points:
Can be used in a for loop
Does not itself keep track of iteration state
Can create an iterator

Example:
my_list = [1, 2, 3, 4]

for item in my_list:
    print(item)

2. What is an Iterator?
An iterator is an object that represents a stream of data. It keeps track of the current position during
iteration and returns elements one at a time using the __next__() method.

Key Points:
Remembers its state
Returns one element at a time
Raises StopIteration when exhausted

Example:

numbers = [10, 20, 30]
it = iter(numbers)

print(next(it))
print(next(it))
print(next(it))

3. Relationship Between Iterable and Iterator
An iterable becomes an iterator when passed to the iter() function. The iterator then produces values using the next() function.

Example:
text = "Python"
iterator = iter(text)

print(next(iterator))
print(next(iterator))

4. Difference Between Iterable and Iterator
In Paragraph Form

An iterable is a container that holds elements and can be iterated over, but it does not track the iteration process. An iterator, on the other hand, is responsible for performing the iteration and remembering the current position. An iterable can create multiple iterators, whereas an iterator cannot be reset once it is exhausted.

Key Differences (Bullets):

Iterable - Can generate an iterator
Iterator - Produces values using next()
Iterable - Can be reused
Iterator - Single-use, gets exhausted

Example:
nums = [1, 2, 3]

it1 = iter(nums)
it2 = iter(nums)   # Another iterator from same iterable

5. Custom Iterator Using a Class
Python allows users to create their own iterators by defining the __iter__() and __next__() methods in a class.

Example:
class Count:
    def __init__(self, limit):
        self.num = 1
        self.limit = limit

    def __iter__(self):
        return self

    def __next__(self):
        if self.num <= self.limit:
            val = self.num
            self.num += 1
            return val
        else:
            raise StopIteration

for i in Count(3):
    print(i)

6. Importance of Iterators
Iterators are memory-efficient because they generate values one at a time instead of storing all values at once.
This makes them useful when working with large datasets.

Example:
for i in range(1000000):
    pass   # range() returns an iterator-like object


6.Explain the concept of generators in Python and how they are defined.
Ans. 1. What is a Generator?
A generator in Python is a special type of function that produces values one at a time using the yield keyword
instead of returning all values at once. Unlike normal functions, generators pause their execution after yielding a value and resume from the same point when the next value is requested.

Key Points:
Uses yield instead of return
Produces values lazily (on demand)
More memory-efficient than lists

Example:
def my_generator():
    yield 1
    yield 2
    yield 3

2. How Generators Work
When a generator function is called, it does not execute immediately. Instead, it returns a generator object, which
can be iterated over using next() or a for loop. Each call to next() resumes execution until the next yield statement.

Example:
gen = my_generator()
print(next(gen))
print(next(gen))
print(next(gen))

3. Defining a Generator Function
A generator function is defined like a normal function but contains one or more yield statements. Each yield sends
a value to the caller and pauses the function’s state.

Key Points:
Defined using def
Must contain yield
Automatically implements iterator protocol

Example:
def count_up(n):
    i = 1
    while i <= n:
        yield i
        i += 1

4. Using Generators with a for Loop
Generators are commonly used with for loops, which automatically handle the iteration and stop when the generator is exhausted.

Example:
for num in count_up(5):
    print(num)

5. Generator Expression
A generator expression is a compact way to create a generator, similar to list comprehensions but using parentheses instead of square brackets.

Key Points:
Short and concise
Memory-efficient
Evaluated lazily

Example:
squares = (x * x for x in range(5))
for value in squares:
    print(value)


7. What are the advantages of using generators over regular functions?
Ans. Generators provide several benefits compared to regular functions, especially when dealing with large datasets
or continuous data streams.

1. Memory Efficiency
Generators are memory-efficient because they generate values one at a time instead of storing all values in memory at once.
Regular functions usually return complete data structures like lists, which consume more memory.

Key Points:
Values are produced on demand
Suitable for large datasets

Example:
def gen_numbers(n):
    for i in range(n):
        yield i

# Uses less memory
for num in gen_numbers(1000000):
    pass

2. Lazy Evaluation
Generators use lazy evaluation, meaning values are computed only when needed. Regular functions compute
all results immediately before returning them.

Key Points:
Improves performance
Avoids unnecessary computation

Example:
def squares(n):
    for i in range(n):
        yield i * i

3. Better Performance for Large Data
Since generators do not store all values in memory, they are faster and more efficient when handling large amounts
of data compared to regular functions that return lists.

Example:
# Regular function
def square_list(n):
    return [i * i for i in range(n)]

# Generator
def square_gen(n):
    for i in range(n):
        yield i * i

4. Ability to Represent Infinite Sequences
Generators can produce infinite sequences, whereas regular functions cannot do this safely due to memory limitations.

Key Points:
No memory overflow
Useful for streams and continuous data

Example:
def infinite_numbers():
    n = 1
    while True:
        yield n
        n += 1

5. Simplified Code for Iteration
Generators simplify complex iteration logic by maintaining state automatically. This makes the code cleaner
compared to regular functions using loops and lists.

Example:
def countdown(n):
    while n > 0:
        yield n
        n -= 1

6. Automatic State Management
Generators automatically remember their execution state between calls. Regular functions start execution from
the beginning each time they are called.

Key Points:
No need for manual state tracking
Easy to pause and resume

Example:
gen = countdown(3)
print(next(gen))
print(next(gen))

7. Integration with Iteration To
Generators work seamlessly with Python’s iteration tools like for loops, sum(), max(), and list().

Example:
total = sum(x for x in range(10))
print(total)


8. What is a lambda function in Python and when is it typically used?
Ans. 1. What is a Lambda Function?
A lambda function in Python is a small, anonymous (nameless) function defined using the lambda keyword.
Unlike regular functions, lambda functions are written in a single line and can have any number of arguments but only one
expression. The expression is evaluated and automatically returned.

Key Points:
Anonymous function (no name)
Written in one line
Uses lambda instead of def

Example:
square = lambda x: x * x
print(square(5))

2. Syntax of a Lambda Function
The syntax of a lambda function is simple and compact.

Syntax:
lambda arguments: expression

Example:
add = lambda a, b: a + b
print(add(3, 4))

3. How Lambda Functions Work
When a lambda function is executed, it evaluates the given expression and returns the result automatically without
using the return keyword.

Example:
maximum = lambda a, b: a if a > b else b
print(maximum(10, 20))

4. When Are Lambda Functions Typically Used?
Lambda functions are typically used when a small function is needed for a short period of time, especially as
an argument to another function.

a) Used with map()
Lambda functions are commonly used with map() to apply an operation to each element of an iterable.

Example:
numbers = [1, 2, 3, 4]
squares = list(map(lambda x: x * x, numbers))
print(squares)

b) Used with filter()
Lambda functions are used with filter() to select elements based on a condition.

Example:
numbers = [1, 2, 3, 4, 5]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)

c) Used with sorted()
Lambda functions are often used to define custom sorting logic.

Example:
students = [("Ravi", 85), ("Anita", 92), ("Sohan", 78)]
sorted_students = sorted(students, key=lambda x: x[1])
print(sorted_students)

5. Advantages of Lambda Functions
Lambda functions reduce code length and improve readability when the logic is simple.

Key Points:
Less code
No need to define a full function
Useful for short-term operations

Example:
print((lambda x: x + 10)(5))

6. Limitations of Lambda Functions
Lambda functions are limited to a single expression and are not suitable for complex logic.

Key Points:
Only one expression allowed
No statements like loops or assignments

Example (Not Allowed):
# lambda x:
#   for i in range(x):


9.Explain the purpose and usage of the `map()` function in Python.
Ans. 1. What is the map() Function?
The map() function in Python is a built-in function used to apply a given function to each element of an iterable
(such as a list, tuple, or set). It processes elements one by one and returns a map object (an iterator) containing the results.

Key Points:
Applies a function to every item
Works with iterables
Returns an iterator (map object)

Example:
def square(x):
    return x * x
numbers = [1, 2, 3, 4]
result = map(square, numbers)
print(list(result))

2. Purpose of Using map()
The main purpose of map() is to transform data efficiently without using explicit loops. It improves code readability
and reduces the amount of code needed for repetitive operations.

Key Points:
Avoids writing for loops
Makes code concise and readable
Useful for data transformation

Example:
numbers = [1, 2, 3]
doubled = list(map(lambda x: x * 2, numbers))
print(doubled)

3. Syntax of map()
The general syntax of the map() function is:
map(function, iterable)

Where:
function - Operation to apply
iterable - Sequence of elements

Example:
map(int, ["1", "2", "3"])

4. Using map() with Lambda Functions
map() is often used with lambda functions when the operation is simple and does not need a full function definition.

Example:
numbers = [5, 10, 15]
result = map(lambda x: x + 5, numbers)
print(list(result))

5. Using map() with Multiple Iterables
The map() function can take more than one iterable. The function is applied to corresponding elements from each iterable.
 Mapping stops when the shortest iterable ends.

Example:
a = [1, 2, 3]
b = [4, 5, 6]
result = map(lambda x, y: x + y, a, b)
print(list(result))

6. Output of map() Function
The map() function returns a map object, which can be converted into a list, tuple, or set.

Example:
result = map(lambda x: x * 3, [1, 2, 3])
print(tuple(result))

7. Difference Between map() and a Loop
Using map() is often cleaner and more efficient than writing a loop for simple transformations.

Example Using Loop:
result = []
for i in [1, 2, 3]:
    result.append(i * 2)

Example Using map():
result = list(map(lambda x: x * 2, [1, 2, 3]))


10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
Ans. In Python, map(), filter(), and reduce() are higher-order functions used to process iterables efficiently.
Although they may seem similar, each serves a different purpose.

1. map() Function
Purpose
The map() function is used to apply a function to every element of an iterable and return the transformed results.

Key Points:
Transforms data
Returns a map object (iterator)
One-to-one mapping

Example:
numbers = [1, 2, 3, 4]
squares = list(map(lambda x: x * x, numbers))
print(squares)

2. filter() Function
Purpose
The filter() function is used to select elements from an iterable that satisfy a given condition.

Key Points:
Filters data based on condition
Returns elements that evaluate to True
Does not modify values

Example:
numbers = [1, 2, 3, 4, 5]
even_nums = list(filter(lambda x: x % 2 == 0, numbers))
print(even_nums)

3. reduce() Function
Purpose
The reduce() function is used to reduce an iterable to a single value by repeatedly applying a function.

Key Points:
Performs cumulative operation
Returns a single result
Must be imported from functools

Example:
from functools import reduce
numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)

4. Key Differences
The map() function transforms each element of an iterable and returns a new iterable of the same length.
The filter() function selects only those elements that satisfy a condition and returns a possibly smaller iterable.
In contrast, the reduce() function combines all elements of an iterable into a single cumulative value.

5. Summary
map() - Changes each element
filter() - Selects elements
reduce() - Combines elements

Example Comparison:
nums = [1, 2, 3, 4]
print(list(map(lambda x: x * 2, nums)))
print(list(filter(lambda x: x > 2, nums)))
from functools import reduce
print(reduce(lambda x, y: x + y, nums))


11. Using pen & Paper write the internal mechanism for sum operation using  reduce function on this given list:[47,11,42,13];
Ans. See cell below
'''
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Check primes from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(num, "is prime")
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a = 0
        self.b = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.count += 1
            self.a, self.b = self.b, self.a + self.b
            return self.b

# Example usage
fib = Fibonacci(10)
for num in fib:
    print(num)
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

# Example usage
for value in powers_of_two(5):
    print(value)
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.rstrip('\n')  # Remove newline characters

# Example usage
for line in read_file_lines("example.txt"):
    print(line)
```
